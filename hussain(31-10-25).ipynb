```python
import pandas as pd
```

```python
import numpy as np
```

```python
from sklearn import linear_model
```

```python
df=pd.read_csv('homeprices.csv')
```

## Dummies variable using pandas 

```python
df
```

```python
df.town.duplicated()
```

```python
dumiesdf=pd.get_dummies(df.town)
dumiesdf
```

```python
dumiesdf.info()
```

```python
boole=dumiesdf[['monroe township','robinsville','west windsor']].replace({False:0,True:1})
```

```python
merg=pd.concat([df,boole],axis='columns')
merg
```

```python
final=merg.drop(['town','west windsor'],axis='columns')
final
```

```python
y=final.price
```

```python
x=final.drop('price',axis='columns')
x
```

```python
reg=linear_model.LinearRegression()
```

```python
reg.fit(x,y)
```

```python
reg.predict(x).round(2)
```

```python
reg.score(x,y)
```

```python
reg.predict([[3200,1,0]])
```

```python
reg.predict([[5000,1,0]])   #township
```

```python
reg.predict([[5000,0,0]])    #windsor
```
